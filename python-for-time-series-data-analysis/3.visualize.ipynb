```python
import pandas as pd
```

```python
dff["c"].plot.line(figsize=(20, 3), ls='-.')
```

```python
dff["c"].plot.line(figsize=(20, 3), ls='-.', c='red')
```

```python
dff["c"].plot.line(figsize=(20, 3), ls='-.', c='red', lw=5)
```

#### title, xlabel, ylabel

```python
title = "My Plot Title"
xlabel = 'My X Label'
ylabel = "My L Label"
```

```python
ax = dff["c"].plot.line(figsize=(10, 3), ls=":", c='red', lw=5, title=title)
ax.set(xlabel=xlabel, ylabel=ylabel)
```

```python
dff.plot()
```

For starters we can pass a location code. <tt>.legend(loc=1)</tt> places the legend in the upper-right corner of the plot.<br>Alternatively we can pass a location string: <tt>.legend(loc='upper right')</tt> does the same thing.

```python
axx = dff.plot()
axx.legend(loc=2)
```

```python
axx2 = dff.plot()
axx2.legend(loc=2, bbox_to_anchor=(2.0,1.0))
```

```python
axx3 = dff.plot()
axx3.legend(loc=3, bbox_to_anchor=(1.0,1.0))
```
